```python
def convert_(data):
    for i, dialog in enumerate(data):
        dial = []
        for d in dialog['dialog']:
            if d['speaker'] == 'seeker':
                dial.append({
                    'text': d['content'].replace('\n',''),
                    'speaker': 'usr',
                })
            if d['speaker'] == 'supporter':
                dial.append({
                    'text': d['content'].replace('\n',''),
                    'speaker': 'sys',
                    'strategy': d['annotation']['strategy'],
                })
        data[i]['dialog'] = dial
        
        del data[i]['experience_type']
        del data[i]['survey_score']
        del data[i]['seeker_question1']
        del data[i]['seeker_question2']
        del data[i]['supporter_question1']
        del data[i]['supporter_question2']

    return data
```

```python
train = convert_(train_data)
valid = convert_(dev_data)
test = convert_(test_data)
```

```python
test[0]
```

```python
print('train', len(train))
with open('./train.txt', 'w', encoding='UTF-8') as f:
    for e in train:
        f.write(json.dumps(e) + '\n')

print('valid', len(valid))
with open('./valid.txt', 'w', encoding='UTF-8') as f:
    for e in valid:
        f.write(json.dumps(e) + '\n')

print('test', len(test))
with open('./test.txt', 'w', encoding='UTF-8') as f:
    for e in test:
        f.write(json.dumps(e) + '\n')
        
with open('./sample.json', 'w', encoding='UTF-8') as f:
    json.dump(train[:10], f, ensure_ascii=False, indent=2)
```
